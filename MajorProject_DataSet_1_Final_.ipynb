# 2. Preparation and Cleaning of Data

```python
df=df.drop(['dist4'],axis=1).copy()   # Deleting column 'dist4'
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Indentify Null Values

df.isna().sum()    # sum the total null values
```

```python
df.dropna(inplace=True)    # Drop rows with any null values

df.dropna(axis=1, inplace=True)   # Drop columns with any null values

df.isna().sum()
```

```python
# Duplicated data

df.duplicated()    # Finding Duplicated data
```

```python
df.loc[df.duplicated()]  # Locating duplicated data
```

```python
# duplicated columns

df.loc[df.duplicated(subset=['resid_area'])]
```

```python
df.query('resid_area == 37.87')  # Listing rows with 'resid_area == 37.8'
```

```python
df.loc[df.duplicated(subset=['resid_area','price','age'])]   # Locating duplicated rows of subset
```

```python
df.duplicated(subset=['resid_area','price','age']).sum()  # Sum of duplicated rows in subset
```

```python
df.loc[~df.duplicated(subset=['resid_area','price','age'])].reset_index(drop=True)  # Locating Unique rows and re-indexing them without additional index column
```

```python
df=df.loc[~df.duplicated(subset=['resid_area','price','age'])].reset_index(drop=True).copy() # Placing refined data in df
```

# 3. Feature Analysis

```python
df['price'].value_counts()  # Listing prices and count of similar prices
```

```python
ax = df['price'].value_counts().head(10).plot(kind='bar',title='Top 10 Prices')  # Plotting Bar Graph with price and count as mentioned in above cell
ax.set_xlabel("Price")
ax.set_ylabel('Count of house with common price')
```

```python
df['rainfall'].plot(kind='kde',title='Rainfall Density')  # Plotting KDE on rainfall
```

```python
df.plot(kind='scatter',x='resid_area',y='price',title='Residential Area VS Price')   # Plotting Scatter Map on 'resid_area' and 'price'
```

```python
df.plot(kind='scatter',x='age',y='price',title='Age VS Price')  # Plotting Scatter Map on 'age','price'
```

```python
ax = sns.scatterplot(x='resid_area',y='price',hue='age',data=df)   # Plotting Scatter Map on 'resid_area' and 'price' with hue of 'age'
ax.set_title("Residential Area VS Age VS Price")
plt.show()
```

```python
sns.pairplot(df,vars=['resid_area','price','age','room_num','airport'])   # Plotting pairplot on 'resid_area','price','age','room_num' and 'airport'
plt.show()
```

```python
# correlation

df[['dist1','dist2','dist3']].dropna().corr()   # Finding the Correlation in 'dist1','dist2' and 'dist3'
```

```python
cor = df[['price','crime_rate', 'air_qual', 'room_num', 'dist1', 'teachers', 'poor_prop', 'rainfall', 'parks']].dropna().corr()

sns.heatmap(cor,annot=True)   # Plotting Heat Map on 'cor' which is correlation of 'price','crime_rate', 'air_qual', 'room_num','dist1', 'teachers', 'poor_prop', 'rainfall' and 'parks
```

```python
# 2. State the price of area with Highest Crime rate


max_row = df['crime_rate'].idxmax()  # Finding row in 'crime_rate' column

highest_price = df.loc[max_row, 'price']    # Extracting the 'price' from selected row
print(highest_price)
```

```python
# 3. State the residential area of area with Highest air quality


max_row = df['air_qual'].idxmax()  # Finding row in 'air_qual' column

highest_price = df.loc[max_row, 'resid_area']    # Extracting the 'resid_area' from selected row
print(highest_price)
```

```python
# 4. State the price of area with Highest rainfall


max_row = df['rainfall'].idxmax()     # Finding row in 'rainfall' column

highest_price = df.loc[max_row, 'price']    # Extracting the 'price' from selected row
print(highest_price)
```

```python
# 5. State the Crime rate where there is Highest number of parks


max_row = df['parks'].idxmax()     # Finding row in 'parks' column

highest_price = df.loc[max_row, 'crime_rate']     # Extracting the 'crime_rate' from selected row
print(highest_price)
```
